```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("Carrot.jpg")
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
R , G , B = cv2.split(img)
```

```python
cols,rows = G.shape
```

```python
for i in range(cols):
    for j in range(rows):
        if G[i,j] < 219 and R[i,j]>10: 
            R[i,j] = G[i,j]

result = cv2.merge((R, G,B ))
plt.imshow(result)

```

```python
result = cv2.merge((B, G,R ))
cv2.imwrite("result.jpg" , result)
```
