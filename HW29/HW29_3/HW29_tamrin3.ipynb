```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("rubix.png")
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
R , G , B = cv2.split(img)
```

```python
col_g , row_g =G.shape
for i in range(col_g):
    for j in range(col_g):
        if G[i,j]<200 and B[i,j]>150 and R[i,j]>150:
            B[i,j] = 0
            R[i,j] = 0
            G[i,j] = 255
```

```python
for i in range(col_g):
    for j in range(col_g):
        if R[i,j]<200 and B[i,j]>150 and G[i,j]>150:
            B[i,j] = 0
            G[i,j] = 0
            R[i,j] = 255
```

```python
for i in range(col_g):
    for j in range(col_g):
        if B[i,j]<200 and R[i,j]>150 and G[i,j]>150:
            R[i,j] = 0
            G[i,j] = 0
            B[i,j] = 255
```

```python
img_result = cv2.merge((R , G , B))
plt.imshow(img_result)
```

```python
img_result = cv2.merge((B , G , R))
cv2.imwrite("result.jpg" , img_result)
```
