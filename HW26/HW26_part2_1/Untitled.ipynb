```python
import cv2
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("mnist.png" , 0)

def number(rows,cols,num,x) :

        for i in range (x , x+100 ,int((rows/ (num/2)))):
                for j in range(0 , cols , int((cols/num))) :
                        target = img[i:i+int((cols/num)), j:j+int((cols/num))]
                        cv2.imwrite(f'small_image{x}/output{i},{j}.jpg', target)

rows,cols = img.shape
num=rows/10

for k in range (0,1000,100):
        number(rows,cols,num,k)


cv2.imwrite("re.jpg" , img)
plt.imshow(img)
```
