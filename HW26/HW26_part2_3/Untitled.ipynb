```python
import cv2
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("ms.jpg" , 0)
h ,w =img.shape
d = 1 - 0.025
for i in range(h):
    for j in range(w):
        r = random.random()
        if r < 0.025:
            img[i][j] = 0
        elif r > d:
            img[i][j] = 255
        else:
            img[i][j] = img[i][j]
cv2.imwrite("noise_img.jpg" , img)
plt.imshow(img)
```

```python
mask   = np.ones((5 , 5)) /25 
rows , cols = img.shape
result = np.zeros((rows,cols))
for i in range(5 //2 ,rows-(5 //2)):
    for j in range(5//2 ,cols-(5// 2 )):
        small_img = img[i-(5//2):i+1+(5//2) ,j-(5//2):j+1+(5//2)]
        small_img_id = small_img.reshape(25)
        small_img_1d_sorted = np.sort(small_img_id)
        result[i,j] = small_img_1d_sorted[4]
cv2.imwrite("Noiseless_img.jpg" , result)
plt.imshow(result)
```
