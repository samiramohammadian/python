```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_orginal = cv2.imread("0.jpg" )
img= cv2.imread("0.jpg" , cv2.IMREAD_GRAYSCALE)
#_ , img =cv2.threshold(img , 75 , 255 , cv2.THRESH_BINARY)
plt.imshow(img , cmap= "gray")
```

```python
img_canny = cv2.Canny(img,0,255)
plt.imshow(img_canny,cmap="gray")
```

```python
img_thresh=cv2.adaptiveThreshold(img_canny,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,7,5)
kernel = np.ones((9,9),np.uint8)
img_thresh = cv2.dilate(img_thresh,kernel,iterations = 1)
plt.imshow(img_thresh,cmap="gray")
```

```python
contours,_ = cv2.findContours(img_thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
x,y,w,h = cv2.boundingRect(contours[0])
dice = img_thresh[y:y+h,x:x+w]
```

```python
contours,hierarchy = cv2.findContours(dice,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
first_child = hierarchy[0][:,2]>0
count = np.bincount(first_child)[0]
print("Dice Recognition: ",count)
```

```python
cv2.rectangle(img_orginal,(x,y),(x+w,y+h),(0,0,255),4)
cv2.putText(img_orginal,f"Dice Recognition:{count}", (x-4,y-15), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255,0,255),2)
plt.imshow(img_orginal)
```

```python
result_number = cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 1, 20, param1=40, param2=30, minRadius=1, maxRadius=30)
print(result_number)
```

```python
print(len(result_number))
result_number = result_number[0, :]
print(len(result_number))

for circle in result_number:
    print(circle)
    cv2.circle(img_orginal, (int(circle[0]), int(circle[1])), int(circle[2]), (250, 0, 0), 2)
    
plt.imshow(img_orginal)
```
